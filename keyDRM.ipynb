```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_filepath = ".../DRM_Data.csv"
data_keyDRM = pd.read_csv(data_filepath)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=data_keyDRM["total_key"])
```

```python
import matplotlib.dates

# Convert 'date' column to datetime type
data_keyDRM["date"]=pd.to_datetime(data_keyDRM["date"])

# Set the size of the plot
plt.figure(figsize=(30,10))

# Create line chart
sns.lineplot(x="date", y="total_key", data=data_keyDRM)

# Format the x-axis to focus on month issuance
plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%Y-%m"))

# Set interval for date labels
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.DayLocator(interval=5))

# Rotate date labels for better readability
plt.gcf().autofmt_xdate()

# Add axis labels and title
plt.xlabel("Date")
plt.ylabel("Total Key")
plt.title("Digital Right Management status by month")

plt.savefig("keyDRM.png")
plt.show()
```

## Comment on DRM status by month
### Mid 2019: 
The data start with mid-year (from May to August) where the line chart reported around 2000 keys were permitted to users.

### Late 2019 to Early 2020:
From mid-2019, there was a significant drop in total keys count around September 2019. <br>
After this drop, total keys remained very low, almost reaching 0 and gradually increasing in early 2020. Reaching its second highest peaks in April with over 6000 keys were permitted to users.

### Mid 2020:
The Key status continues to increase, especially around May, reaching its highest at the beginning of June with 6000 keys.

## Summary:
The line chart of "Key DRM" pattern can help in planning for the following suggestions: <br>
